Ensuite, on doit trouver quel est le programme qui donne le plus de subventions et créer un sous-ensemble de ce dernier. 
Avec .value_counts(), on se rend compte que c'est le programme "Jeunesse - emplois d'été Canada" qui arrive en premier avec une belle avance. 

```python
frenchsub.nom_prog.value_counts()
```

Je crée donc mon sous ensemble qui va se nommer "bestprog"

```python
bestprog = frenchsub.nom_prog == "Jeunesse - Emplois d'été Canada"
```

Puis je vérifie par la suite si tout fonctionne correctement

```python
frenchsub[bestprog]
```

Je vais ensuite regarder qu'elle est l'entreprise/la société qui a le plus bénéficié de ce programme. Elles seront toutes classées par ordre décroissant.

```python
frenchsub[bestprog].beneficiaire_nom_legal.value_counts()
```

Enfin, je vais regarder quelles sont les provinces qui ont le plus bénéficié de ce programme et classer ces résultats en un pourcentage. 

```python
frenchsub[bestprog].province_beneficiaire.value_counts()
```

On transforme les résultats obtenus en pourcentage (le Québec à la deuxième place !)

```python
frenchsub[bestprog].province_beneficiaire.value_counts() / len(frenchsub[bestprog] * 100)
```
